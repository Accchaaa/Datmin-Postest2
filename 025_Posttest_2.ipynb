# (1) Apriori

Gunakan dataset ini : https://drive.google.com/file/d/1BqC-2d1ZYqFIFwdjdsW5bAAqNFh2AO4C/view?usp=share_link

Dan buatlah daftar support dan confidence untuk 2 item set dengan menggunakan dataset tersebut (hint : buatlah tabel tabulasi terlebih dahulu untuk memudahkan)

````python
df = pd.read_excel ("Tabular.xlsx")
df
````

````python
#2 item set
from itertools import permutations
# Extract unique items
rules = list(permutations(df.columns, 2))
rules
````

````python
rule_baru = []
for rule in rules:

  if len(df[(df[rule[0]]==1) & (df[rule[1]]==1)]) >0:
    rule_baru.append(rule)
rule_baru
````

````python
hitung = {} 
for rule in rule_baru:
  banyak_data= len(df[
      (df[rule[0]]==1) & (df[rule[1]]==1) ])  
  hitung[str(rule)] = banyak_data
hitung
````

````python
#Daftar support
print("---")
for item,ab in hitung.items():
  print(f"{item} support = {ab/len(df)*100}%")
````

````python
# Daftar Confidence

print("---")
for item,ab in hitung.items():
  #"A->B"
  i = eval(item) 
  print(f"{i[0]}->{i[1]}  Confidence = {ab/df[i[0]].sum()*100}%")
````

````python
df = pd.read_csv ("deteksi_uang_palsu.csv")
df
````

````python
nilai_distance = 2

# Tambah satu row sebagai penanda targetnya
target = ['Target ',172.25, 103.94, 103.68, 4.14, 3.25, 112.48]
df.loc[len(df.index)] = target
df
````

````python
# 1. Tambahkan kolom jarak untuk menampilkan jarak dari dataset terhadap target prediksi
import math
jarak = []
for i in range(len(df['is_genuine'])):
  diagonal = df['diagonal'][i] - target[1];
  height_left = df['height_left'][i] - target[2];
  height_right = df['height_right'][i] - target[3];
  margin_low = df['margin_low'][i] - target[4];
  margin_up = df['margin_up'][i] - target[5];
  length = df['length'][i] - target[6];
  jarak.append(math.sqrt(diagonal**nilai_distance + 
                         height_left**nilai_distance +
                         height_right**nilai_distance +
                         margin_low**nilai_distance +
                         margin_up**nilai_distance +
                         length**nilai_distance))

df['Jarak'] = jarak
df
````

````python
# 2. Urutkan data berdasarkan K jarak terdekat (Untuk K nya ambil akar 2 dari banyak data)

# Hapus row target
df.drop(10,axis=0,inplace=True)

# Mencari yang terdekat
K = round(math.sqrt(len(jarak)))
df_urut = df.sort_values('Jarak').head(K)
df_urut
````

````python
# 3. Tentukan apa hasil prediksinya berdasarkan label terbanyak
K2 = round(math.sqrt(len(df['is_genuine'])))
hapuslast= jarak.pop()

index1 = sorted(range(len(jarak)), key=lambda K2: jarak[K2]) [:K2]
closer = [df['is_genuine'][i] for i in index1]
prediction = max(set(closer), key=closer.count)

print('Hasil prediksi berdasarkan label terbanyak: ', prediction)
````

### Business Understanding

Bank harus berusaha keras tidak hanya untuk mendapatkan pelanggan baru, namun juga untuk mempertahankan pelanggan lama. Prediksi nasabah churn adalah kegiatan yang dilakukan untuk memprediksi nasabah tersebut akan meninggalkan perusahaan atau tidak.
Dengan mempelajari data nasabah (dataset customer churn), bank akan melakukan analisis menggunakan teknik data mining tentang apa yang menjadi faktor-faktor penyebab customer churn. Hasil analisis itu akan digunakan bank agar bisa melakukan evaluasi dan meningkatkan layanan-layanan yang ada sehingga customer churn dapat dicegah.

### Data Understanding

Data ini berisi data pelanggan yang menarik rekening mereka dari bank karena beberapa kerugian dan masalah lainnya.

````python
# collect data
df = pd.read_csv ("Churn Modeling.csv")
df
````

````python
#kolom yang ada pada dataset
df.columns
````

Uraian Kolom


*   RowNumber: adalah nomor urut setiap baris pada data
*   CustomerId: nomor Id pelanggan
*   Surname: nama pelanggan
*   CreditScore: merupakan poin yang customer dapat ketika mereka rutin dalam membayar tagihan, mengelola utang, dan pengelolaan keuangan secara umum. Semakin tinggi credit score seseorang, semakin besar kemungkinan mereka untuk mendapatkan kredit dan bunga pinjaman yang lebih rendah.
*   Geography: menunjukan negara tempat tinggal customer
*   Gender: Jenis kelamin customer
*   Age: Usia customer
*   Tenure: sistem kerja yang memberikan jaminan keamanan pekerjaan kepada karyawan bank yang telah bekerja dengan waktu yang cukup lama. Para karyawan yang telah mendapat status tenure, diberikan hak-hak tertentu dan tidak dapat dipecat dengan mudah, kecuali atas alasan yang jelas dan sah.
*   Balance: kestabilan customer
*   NumOfProducts: nomor kategori produk/layanan bank
*   HasCrCard: merujuk pada apakah customer memiliki kartu kredit
*   IsActiveMember: merujuk pada apakah customer aktif dalam melakukan layanan bank
*   EstimatedSalary: estimasi gaji
*   Excited: apakah customer gembira atau antusias
